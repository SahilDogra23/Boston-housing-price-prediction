# Finance Agent with Agno

This notebook demonstrates how to build an AI-powered financial analysis system using Agno. It combines web searching capabilities and financial data analysis tools to create a comprehensive investment research platform. The system utilizes OpenAI's GPT model along with specialized tools for market data retrieval and analysis.

- Creates a web agent for general internet research using DuckDuckGo integration
- Implements a dedicated finance agent with YFinance tools for stock analysis and market data
- Combines both agents into a team for enhanced financial research capabilities

### Creating Agents for Financial Analysis

We'll create specialized agents using Phidata - a web agent for general research and a finance agent for market data analysis. These agents will be combined into a team to provide comprehensive financial insights.


```python
from agno.agent import Agent
from agno.team import Team
from agno.models.openai import OpenAIChat
from agno.tools.duckduckgo import DuckDuckGoTools
from agno.tools.yfinance import YFinanceTools



# Create Web Research Agent
web_agent = Agent(
    name="Web Research Agent",
    role="Search the web for financial news, market trends, and general information",
    model=OpenAIChat(id="gpt-4o-mini"),
    tools=[DuckDuckGoTools()],
    instructions=[
        "You are a web research specialist focused on financial and market information.",
        "Always provide sources for your information.",
        "Focus on recent and relevant financial news and trends."
    ],
    add_history_to_context=True,
    markdown=True,
)

# Create Finance Data Agent
finance_agent = Agent(
    name="Finance Data Agent",
    role="Analyze financial data, stock prices, and provide investment insights",
    model=OpenAIChat(id="gpt-4o-mini"),
    tools=[
        YFinanceTools(
            include_tools=[
              "get_current_stock_price",
                "get_company_info",
                "get_historical_stock_prices",
                "get_company_news",
            ]
        )
    ],
    instructions=[
        "You are a financial analysis expert specializing in stock market data.",
        "Always use tables to display financial data for better readability.",
        "Provide comprehensive analysis including technical and fundamental metrics.",
        "Include risk assessments and investment recommendations when appropriate."
    ],
    add_history_to_context=True,
    markdown=True,
)
```

```python
from agno.agent import Agent
from agno.models.openai import OpenAIChat
from agno.tools.duckduckgo import DuckDuckGoTools
from agno.tools.yfinance import YFinanceTools

web_agent = Agent(
    name="Web Agent",
    role="Search the web for information",
    model=OpenAIChat(id="gpt-4o-mini"),
    tools=[DuckDuckGoTools()],
    markdown=True,
)

finance_agent = Agent(
    name="Finance Agent",
    role="Get financial data",
    model=OpenAIChat(id="gpt-4o-mini"),
      tools=[
        YFinanceTools(
            include_tools=[
              "get_current_stock_price",
                "get_company_info",
                "get_historical_stock_prices",
                "get_company_news",
            ]
        )
    ],

    instructions=["Always use tables to display data"],
    markdown=True,
    )
```

```python
# Create Financial Analysis Team using Agno v2.0 Team class
financial_team = Team(
    name="Financial Analysis Team",
    role="Comprehensive financial research and analysis team",
    members=[web_agent, finance_agent],
    instructions=[
        "You are leading a team of financial research specialists.",
        "Coordinate between web research and financial data analysis.",
        "Provide comprehensive, well-researched financial insights.",
        "Always cross-reference web research with actual financial data.",
        "Present findings in a clear, structured format with actionable insights."
    ],
    add_history_to_context=True,
    markdown=True,
)
```

### Simple Query

```python
# Test the financial team with a simple query
financial_team.print_response(
    "How much did Nvidia stock increase in the last 1 year? Include both the percentage change and recent news about the company.",
    stream=True
)
```

### Comparing Stocks

```python
def compare_stocks(agent, symbol1, symbol2, period='1y'):
    query = f"Compare the performance metrics of {symbol1} and {symbol2} over the last {period}. Do a detailed analysis on P/E ratio"
    response = financial_team.print_response(
        query,
        stream=True,
        show_full_reasoning=True
    )
    return response
```

```python
# Compare two stocks
compare_stocks(finance_agent, "NVDA", "TSLA", "1y")
```

### Portfolio Recommendations

```python
# Create a dummy portfolio as a string
portfolio_details = """
Portfolio:
1. AAPL (Apple Inc.): 100 shares
2. MSFT (Microsoft Corporation): 50 shares
3. GOOGL (Alphabet Inc.): 30 shares
4. AMZN (Amazon.com, Inc.): 20 shares

Please analyze this portfolio and provide:
1. A brief overview of each stock's recent performance
2. An analysis of the portfolio's diversification
3. Recommendations for potential changes or additions to the portfolio
4. Any relevant market trends that might impact this portfolio

I am ready to take only moderate risk. I want to exit the portfolio in 5 years. Tune your suggestions accordingly.

"""

# Use the existing agent_team to analyze the portfolio
financial_team.print_response(portfolio_details, stream=True)
```
